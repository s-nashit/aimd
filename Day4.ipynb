```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
text = [
    'Apple is launching new iphone in July.',
    'Motorola is launching new Edge in July.',
    'Amazon is launching a new platform  in July.'
    'I will leave in July.',
    'I am eating chocolates.',
    'I am eating Biryani and you are eating rice.'
]
```

```python
text = [
    'I am happy',
    'I am sad',
    'I feel good',
    'This is terrible and sad',
    'This is worse'
]
```

```python
v = TfidfVectorizer()
```

```python
v.fit(text)
```

```python
trnsformed_op = v.transform(text)
```

```python
trnsformed_op
```

```python
print(v.vocabulary_)
```

```python
f = v.get_feature_names_out()
```

```python
for i in f:
    idx = v.vocabulary_.get(i)
    # print(idx)
    idf_score = v.idf_[idx]
    print(f'{i} : {idf_score}')
```

```python
import pandas as pd
df = pd.read_csv('Ecommerce_data.csv')
```

```python
df.shape
```

```python
df['label'].value_counts()
```

```python
df['Labels'] = df['label'].map({
    'Household' : 0,
    'Electronics' : 1,
    'Clothing & Accessories' : 2,
    'Books' : 3
})
```

```python
df.head(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df.Text,
    df.Labels
)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
model = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('Model',  MultinomialNB())
])
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
classification_report(y_test, y_pred)
```

```python
x_test.head(1)
```

```python
test = ['I was expecting camel brown, as shown in the photograph, because I used this same sandal for 7 years.', 'Great value for money. Extended battery life.']
```

```python
model.predict(test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe = Pipeline([
    ('vect', TfidfVectorizer()),
    ('Model' , RandomForestClassifier())
])
```

```python
pipe.fit(x_train, y_train)
```
